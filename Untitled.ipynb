```python
def download_files(varname, place):
    
    globvar = glob.glob(f'output/12000*{varname}')
    runs = ["_".join(val.split('um')[0].split('_')[2:][:-1]) for val in globvar]

#     places = [val.split('_gia')[0].split('_')[-1] for val in globvar][0]
    modelrun = ["_".join(val.split('_f')[0].split('_')[2:]) for val in globvar]
    ds = xr.open_mfdataset(globvar, concat_dim='modelrun', combine='nested')
    
    try:
        ds = ds.rename({'__xarray_dataarray_variable__':'var'})
        ds = ds.assign_coords({'modelrun':modelrun, 'run':runs, 'place':place})
        
    except:
        pass
    return ds
place = 'fennoscandia'
names = ['giaprior', 'gp_variance', 'da_zp', 'posterior']
ds_giaprior, ds_var, ds_zp, ds_post  = [download_files(it, place) for it in names]
ds_var
```
